```python
#import needed packages
import pandas as pd

#imports our csv data to a variable
air_data = pd.read_csv('AirQuality_Daily_StudentVersion.csv')

#turn our data into a dataframe
air_data = pd.DataFrame(air_data)


```

```python
#task 1: find the 5 locations with the highest mean and median values of each particulate
#groups the data by location
locations = air_data.groupby('sensor.name')

#prints tables for the top 5 mean and median of each particulate
locations.agg(mean_voc = ('voc', 'mean')).sort_values('mean_voc', ascending = False).head(5)
```

```python
locations.agg(median_voc = ('voc', 'median')).sort_values('median_voc', ascending = False).head(5)
```

```python
locations.agg(mean_pm2_5 = ('pm2.5_atm', 'mean')).sort_values('mean_pm2_5', ascending = False).head(5)
```

```python
locations.agg(median_pm2_5 = ('pm2.5_atm', 'median')).sort_values('median_pm2_5', ascending = False).head(5)
```

```python
locations.agg(mean_pm10_0 = ('pm10.0_atm', 'mean')).sort_values('mean_pm10_0', ascending = False).head(5)
```

```python
locations.agg(median_pm10_0 = ('pm10.0_atm', 'median')).sort_values('median_pm10_0', ascending = False).head(5)
```

```python
#task 2: find maxiumim values of particulates as well as when and where they occured

#reduces data to the date, sensor name, and voc columns and then sorts to find the highest value of voc
air_data[['date', 'sensor.name', 'voc']].sort_values('voc', ascending = False).head(1)
```

```python
#task 3:determine the impact temperature and humidity have on air quality
#defines a function that categorizes temperatures into 4 categories 
def temperature(temp):
    if temp < 32:
        return 'Freezing'
    elif temp < 50:
        return 'Cool'
    elif temp < 70:
        return 'Warm'
    else:
        return 'Hot'

#defines a function that categorizes humidity into 3 categories
def humidity(hum):
    if hum < 50:
        return 'Low'
    elif hum < 80:
        return 'Medium'
    else:
        return 'High'

```

```python
#applies the temperature and humidity functions to add columns to the data
air_data['temperature_level'] = air_data['temperature'].apply(temperature)
air_data['humidity_level'] = air_data['humidity'].apply(humidity)
```

```python
#groups data by temperature level
temperature_level = air_data.groupby('temperature_level')

#groups data by humidity level
humidity_level = air_data.groupby('humidity_level')

#provides mean and median for each temperature level
temperature_level.agg(mean_voc = ('voc', 'mean'), median_voc = ('voc', 'median'))
```

```python
temperature_level.agg(mean_2 = ('pm2.5_atm', 'mean'), median_2 = ('pm2.5_atm', 'median'))
```

```python
temperature_level.agg(mean_10 = ('pm10.0_atm', 'mean'), median_10 = ('pm10.0_atm', 'median'))
```

```python
humidity_level.agg(mean = ('voc', 'mean'), median = ('voc', 'median'))
```

```python
humidity_level.agg(mean = ('pm2.5_atm', 'mean'), median = ('pm2.5_atm', 'median'))
```

```python
humidity_level.agg(mean = ('pm10.0_atm', 'mean'), median = ('pm10.0_atm', 'median'))
```
